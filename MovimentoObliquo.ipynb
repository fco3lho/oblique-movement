```python
from vpython import *
import random 
import math

ball = sphere(pos = vector(0,0,0), radius = 0.7, color = color.red)
ground = box(pos = vector(0,0,0), size = vector(100,0.5,3), color = color.yellow)

g2 = graph(width=600, height=300, xtitle='Tempo (s)', ytitle='Altura (m)')
g3 = graph(width=600, height=300, xtitle='Velocidade em y (m/s)', ytitle='Altura (m)')
g4 = graph(width=600, height=300, xtitle='Tempo (s)', ytitle='Posição em x (m)')
graph2 = gcurve(graph = g2, color = color.blue)
graph3 = gcurve(graph = g3, color = color.black)
graph4 = gcurve(graph = g4, color = color.red)

attach_trail(ball)

#User

VelocidadeInicial = 20 #Velocidade inicial em m/s
incVi = ((VelocidadeInicial * 1.02) - (VelocidadeInicial * 0.98))/(2 * math.sqrt(3))
print('Velocidade inicial: (', round(VelocidadeInicial,3), '±', round(incVi, 3),') m/s')

AngI = 60 #Angulo do lançamento em graus
incAng = ((AngI * 1.02) - (AngI * 0.98))/(2 * math.sqrt(3))         
print('Ângulo: (', round(AngI,3), '±', round(incAng, 3),') °')

print('\n')

#Variables

erroV = random.uniform(-incVi,incVi)
Vi = VelocidadeInicial + erroV #Velocidade inicial com incerteza aplicada
print('Velocidade inicial com incerteza aleatória:', round(Vi, 3), 'm/s.')

erroA = random.uniform(-incAng, incAng)
Angulo = AngI + erroA #Angulo do lançamento com incerteza aplicada
print('Ângulo com incerteza aleatória:', round(Angulo, 3), '°.')

print('\n')

#Initial conditions

theta = radians(Angulo)
g = vector(0,-9.8,0)
ball.v = vector(Vi*cos(theta),Vi*sin(theta),0)
t = 0
dt = 0.001 #Intervalo de tempo para atualização do objeto

#Equations

while ball.pos.y >= 0:
    rate(350) #Tempo de intervalo da animação para visualização
    ball.v = ball.v + g*dt # a=Δv/Δt, onde a=g e Δt=dt, sendo Δv=g*dt -> [v = vi + g*dt]
    ball.pos = ball.pos + ball.v*dt # v=Δs/Δt, onde Δt=dt, sendo Δs = v*dt -> [s = si + v*dt]
    
    t = t + dt
    graph2.plot(t, ball.pos.y)
    graph3.plot(ball.v.y, ball.pos.y)
    graph4.plot(t, ball.pos.x)
    
    if (ball.v.y < 0.01) and (ball.v.y > (-0.01)):
        print('A altura máxima atingida pela bola foi:', round(ball.pos.y,2), 'm.')

print('O alcance da bola em relação ao ponto inicial foi: ', round(ball.pos.x,2), 'm.')        
print('O tempo de voo da bola foi: ', round(t,2), 's.')
```
